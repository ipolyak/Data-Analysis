# Подготовка данных и kNN

```python
data['class'].dtype.name
```

Выделим категориальные и числовые характеристики. В качестве категориальной характеристики будет выступать 'class'.

```python
categorical_columns = [c for c in data.columns if data[c].name == 'class']
numerical_columns = [c for c in data.columns if data[c].name != 'class']
print(categorical_columns)
print(numerical_columns)
```

Выявим, есть ли пропущенные значения:

```python
data[numerical_columns].count(axis = 0)
```

Оказалось, что пропущенных значений нет.

```python
nonbinary_columns = categorical_columns
print(nonbinary_columns)
```

```python
data['class'].unique()
```

Проведем векторизацию категориального признака ('class'):

```python
data_nonbinary = pd.get_dummies(data[nonbinary_columns])
print(data_nonbinary.columns)
```

Нормализуем числовые характеристики:

```python
data_numerical = data[numerical_columns]
data_numerical = (data_numerical - data_numerical.mean(axis = 0)) / data_numerical.std(axis = 0)
```

```python
data_numerical.describe()
```

Соединим всё в одну таблицу:

```python
data = pd.concat((data_numerical, data_nonbinary), axis = 1)
print(data.shape)
```

```python
data.describe()
```

Выбрасываем столбец 'class', т.к. мы будем предсказывать это значение

```python
X = data.drop(('class'), axis = 1)
y = data['class']
feature_names = X.columns
```

```python
type(feature_names)
```

```python
feature_names
```

```python
print(X.shape)
print(y.shape)
N, d = X.shape
```

```python
type(X)
```

```python
type(y)
```

Разделим выборку на обучающую и тестовую. 30% идет в тестовую выборку:

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1)

N_train, _ = X_train.shape 
N_test,  _ = X_test.shape 

print(N_train, N_test)
```

Произведем обучение классификатора на тренировочной выборке

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 4)
knn.fit(X_train, y_train)
```

```python
knn
```

Выполняем предсказание и подсчет средней ошибки

```python
y_train_predict = knn.predict(X_train)
y_test_predict = knn.predict(X_test)

err_train = np.mean(y_train != y_train_predict)
err_test  = np.mean(y_test  != y_test_predict)

print(err_train, err_test)
```

Ошибку возможно посчитать ещё одним способом:

```python
err_test = 1 - knn.score(X_test, y_test) # Другой способ
err_test
```

Наглядное представление работы классификатора обеспечивается методом confusion_matrix. Размер данной матрицы определяется количеством различных классов. Элемент (i,j) - число объектов, которые относятся к классу i, но классификатор отнес их к классу j

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test, y_test_predict))
```

Минимизируем ошибку путем подбора параметра k:

```python
from sklearn.model_selection import GridSearchCV
nnb = [1, 3, 5, 10, 15, 20, 25, 35, 45, 55]
knn = KNeighborsClassifier()
grid = GridSearchCV(knn, param_grid = {'n_neighbors': nnb}, cv=10)
grid.fit(X_train, y_train)

best_cv_err = 1 - grid.best_score_
best_n_neighbors = grid.best_estimator_.n_neighbors
print(best_cv_err, best_n_neighbors)
```

Теперь проведем классификацию при k = 55

```python
knn = KNeighborsClassifier(n_neighbors = best_n_neighbors).fit(X_train, y_train)

err_train = np.mean(y_train != knn.predict(X_train))
err_test  = np.mean(y_test  != knn.predict(X_test))

print(err_train, err_test)
```

Значение ошибки уменьшилось. Ошибка предсказания сорта вина стала равной ~0.018
